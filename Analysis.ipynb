```python
countries=data["Reviewer_Nationality"].value_counts() * 100
```

```python

%matplotlib inline

plt.title("Checking percentage of tourists visiting hotels in Europe from other countries")

countries=countries.head(15)
countries.plot(kind='pie',figsize=(40,10),autopct="%1.1f%%")
plt.show()
```

```python
type_visitors = [" Couple "," Family with older children "," Group "," Solo Traveler "]

'''Popularity measured by counting each hotel's family type and sorting the counts '''
inp = input("Enter your family type to find top ten hotels  ")
res=data[data["Tags"].str.contains(inp)].groupby('Hotel_Name')["Hotel_Address"].count().sort_values(ascending=False).head(10)


pd.Series([1,2,3], index=['a','b','c']).plot.bar(figsize=(20,10))
plt.xlabel('Hotel_Name')
plt.ylabel('Family Type')
plt.title('Top 10 Popular Hotels for the Family Type {}'.format(inp))
res.plot(kind = 'bar')
plt.show()
```

```python
corr_df = data[data.columns]    
corr_df = corr_df.corr()

f,ax = plt.subplots(figsize=(12,9))

sns.heatmap(corr_df, cmap='inferno', annot=True)
```
